```python
def sim_match(pairing, team_elo_dict):
    "Match Simulator based on Elo"
    elo1 = team_elo_dict[pairing[0]]
    elo2 = team_elo_dict[pairing[1]]

    m = elo1 - elo2
    prob = 1/(1+10**(-m/400))

    draw_prob = mymodel(prob)
    win_prob = prob/(1-draw_prob)

    result = random()

    if win_prob - result > 0.00001:
        score1 = 1
        score2 = 0
        result = pairing[0]
    elif win_prob + draw_prob - result > 0.00001:
        score1 = 0.5
        score2 = 0.5
        result = 'D'
    else:
        score1 = 0
        score2 = 1
        result = pairing[1]

    to.reportMatch(table, [score1, score2, result])
```

```python
def league_table(tournament):
    league_table = pd.DataFrame(columns = ['ID','Name','Opponents','Results','Points','OMW%', 'Fixed Seating'])

    for player in to.playersDict:
        player_data = list(to.playersDict[player].values())
        player_data.insert(0, player)
        league_table.loc[len(league_table)] = player_data

    return league_table.sort_values(by=['Points','OMW%'], ascending = False).reset_index(drop=True)
```

```python
def sim_season(season_length, teams):
    # Preparing Teams
    teams_shuffled = teams.sample(frac = 1).reset_index(drop=True)
    team_elo_dict = {}
    for row in teams_shuffled.iterrows():
        team = row[0]
        elo = int(row[1].Elo)
        team_elo_dict[team] = elo
    
    # Preparing the Tournament
    to = Tournament()
    for row in teams_shuffled.iterrows():
        name = row[1].Club
        to.addPlayer(row[0], name)
    
    for n in range(season_length):
        pairings = to.pairRound()
        for table in pairings:
            pairing = pairings[table]
            if not type(pairings[table]) is str:
                sim_match(pairing, team_elo_dict)
    return to
```
